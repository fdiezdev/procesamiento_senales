## Ejercicio 3:

Construya 3 vectores columna de 20 elementos cada uno, de modo que la diferencia entre
dos elementos consecutivos sea 0.4 para el primero, 0.6 para el segundo y 2 para el tercero.    
a) Construya una matriz de 20 filas por 3 columnas con los vectores definidos anteriormente.

```python
#Construímos los 3 vectores 

v1 = np.arange(1,1+0.4*20,0.4)
v2 = np.arange(1,1+0.6*20,0.6)
v3 = np.arange(1,1+2*20,2)

#Prueben hacer lo mismo con otras formas de construir vectores

print('vector 1 ','\n', v1,'\n', 'de orden', v1.shape)
print('vector 2 ','\n', v2,'\n', 'de orden', v2.shape)
print('vector 3 ','\n', v3,'\n', 'de orden', v3.shape)

#armamos una matriz M con 3 columnas y 20 filas. Ubicamos los vectores generados en cada columna 

M = np.zeros((20,3))
M[:,0]=v1
M[:,1]=v2
M[:,2]=v3
print('Matriz M construida con vectores v1, v2 y v3', '\n', M)
```

 b) Halle el máximo de la matriz y ubique en qué posición se encuentra.

```python
maxM = np.amax(M)
print('El máximo de la matriz M es', maxM)
posicionMax = np.where(M == maxM) #primer elemento es la posición en filas, segundo elemento posición en columnas. Siempre tener en cuenta que la indexación de python empieza en 0
print('Se encuentra en la fila', posicionMax[0]+1, 'columna', posicionMax[1]+1) 
```

Y el mínimo? Cómo lo encuentro? Y su ubicación? Hay un solo mínimo?

```python
minM = np.amin(M)
print('El mínimo de la matriz M es', minM)
posicionMin = np.where(M == minM) #primer elemento es la posición en filas, segundo elemento posición en columnas. Siempre tener en cuenta que la indexación de python empieza en 0
print('Se encuentra en las filas', posicionMin[0]+1, 'columnas', posicionMin[1]+1, 'respectivamente')
```

# **Ejercicio 4: Tipo de Datos**

* **int:** enteros: 
>>**int8:** Enteros con signo en el rango [-128;127] (8 bits)              
>>**int16:** Enteros con signo en el rango [-32768;32767] (16 bits)   
>>**uint8:** Enteros sin signo en el rango [0;255] (8 bits) 256=2^8  
>>**uint16:** Enteros sin signo en el rango [0;65535] (16 bits)   
>>>>sin signo implica que ---> no vamos a obtener números negativos 
* **float:** coma flotante, decimales        
* **bool:** booleanos (true, false)        
* **complex:** complejos
