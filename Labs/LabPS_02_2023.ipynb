## **Ejercicio 5** 

- Señales discretas con N muestras. Las pensamos periódicas y extendidas con período N muestras. 
- Ahora voy a poder estudiar frecuencias o señales con frecuencia hasta N/2 oscilaciones. 
- $\frac{N}{2}$ es la frecuencia de muestreo de Nyquist de señales muestredas con N muestras.
- El $ k $ me va a indicar cuantas veces oscila la señal
- $S_k$ serían los armónicos de una señal

Demuestre que el conjunto de señales discretas $\left \{ s_k\left [ n \right ]=exp \left (j\frac{2\pi}{N} kn\right ) \right \}_{k=0,.., (N-1)} $. es ortogonal repecto al producto interno canónico para funciones a tiempo discreto periódicas de período $N$

```python
N=50 # cantidad de muestras
c1=complex(0,1) # es el j (complejo) j/j**2=-1
print(c1**2)
n=np.arange(N)
```

```python
k=3 #número entre 0 y N-1, frecuencia, cantidad de oscilaciones

s_k=np.exp(2*c1*np.pi/N*k*n) #k-ésima señal en la base
```

```python
r=28 #número entre 0 y N-1, frecuencia, cantidad de oscilaciones

s_r=np.exp(2*c1*np.pi/N*r*n) #

print('producto interno entre s_'+str(k)+' y s_'+str(r)+':')
print(np.round(np.inner(s_k,s_r),decimals=3))
```

```python
plt.figure(figsize=(15,8))#graficamos la parte real de las señales s_k y s_j
plt.subplot(211)
plt.plot(n, np.real(s_k), color='blue')
plt.title('Parte real de señal exp(j2pikn/N) con k='+str(k)+' oscilaciones en N='+str(N)+' muestras')
plt.subplot(212)
plt.plot(n, np.real(s_r), color='blue')
plt.title('Parte real de señal exp(j2pirn/N) con r='+str(r)+' oscilaciones en N='+str(N)+' muestras')
plt.plot()
```

Si ya muestree FNYSQT es la mitad (caso de senales discretas la FQNYSQT es la mitad)
Es el doble en caso de que no hayamos muestreado (caso de senales continuas la FQNYSQT es el doble).
