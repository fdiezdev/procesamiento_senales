```python
# Importo librerías necesarias
import cv2 as cv2 #para imágenes, Open CV (Computer Vision)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# llevo a la memoria
img = cv2.cvtColor(cv2.imread('./color.png'), cv2.COLOR_BGR2RGB) # esto es para que los canales RGB esten en orden
# graficar la imagen
plt.imshow(img)

# tamaño de la imagen
print(img.shape)
```

```python
# graficamos las tres capas por separado
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

plt.figure(figsize=(16,8))

plt.subplot(131)
plt.title("Capa Red")
plt.imshow(R, cmap='Reds')
plt.colorbar()

plt.subplot(132)
plt.title("Capa Green")
plt.imshow(G, cmap='Greens')
plt.colorbar()

plt.subplot(133)
plt.title("Capa Blue")
plt.imshow(B, cmap='Blues')
plt.colorbar()

plt.show()
```
