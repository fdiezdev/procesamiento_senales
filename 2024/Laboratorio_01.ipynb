## Nyquist de muestreo
En el laboratorio trabajamos con señales muesteadas y ahí aparece la **Frecuencia de Nyquist de muestreo**
Dada la frecuencia de muestreo Fm, se define la frecuencia de Nyquist de muestreo como:
Nm=Fm/2

Ejemplo: si la frecuencia de muestreo es de 200 Hertz, la de Nyquist de muestreo es de 100 Hertz (quiere decir que la máxima frecuencia que podremos observar en nuestra señal es 100 Hertz. Si la señal oscilaba más rápido con el muestreo no lo captamos y lo perdimos)

$F_{muestreo} \geq 2*F_{max}$

$F_{muestreo}/2 \geq F_{max}$

b) Digitalizamos y graficamos la señal con 600 muestras en el intervalo $[0,3)$

```python
M=600
Fm=M/3 #frecuencia de muestreo en Hertz, prueben con menos muestras
#M muestras en 3 segundos
Nm=Fm/2 #Frecuencia de Nyquist muestral
t=np.linspace(0,3,M, endpoint=False)
print('La frecuencia de muestreo es de ',Fm, 'Hertz')
print('La frecuencia de Nyquist del muestreo es',Nm, 'Hertz')
#señal con comp de frec 10 y 13, con amplitudes 3 y 2 respectivamente
s=3*np.cos(5*2*np.pi*t)+2*np.cos(13*2*np.pi*t)
```

```python
#Visualizamos con plot en función del tiempo
plt.figure(figsize=(16,5))
plt.plot(t,s, label='s(t)') #en función del tiempo
plt.xlabel('t (muestreado)')
plt.legend(loc='upper right')
plt.show()
```

```python
#Visualizamos con plot en función del nro de muestra
plt.figure(figsize=(16,5))
plt.stem(s, label='s(t[n])') #en función del nro de muestra
plt.xlabel('muestras')
plt.legend(loc='upper right')
plt.show()
```

## Ej. 7
Defina una señal teórica con dos componentes de frecuencias: 20 y 50 Hertz con amplitudes  7 y 5 respectivamente. Digitalícela en el intervalo [3,4) con frecuencia de muestreo de 10.000 Hertz. Calcule Período de muestreo y Frecuencia de Nyquist del muestreo.

$x(t) = x_1(t)+x_2(t)$

$x_1(t) = 7*cos(20*2 \pi *t )\implies x_1(t) = 7*cos(40 \pi *t )$

$x_2(t) = 5*cos(50*2 \pi *t )\implies x_2(t) = 5*cos(100 \pi *t )$

Entonces:

$w_1 = 2 \pi f_1 \implies f_1 = 40\pi / 2\pi = 20[Hz]$

$w_2 = 2 \pi f_2 \implies f_2 = 100\pi / 2\pi = 50[Hz]$


```python
# intervalo
inicio = 3
fin = 4
Fm = 10000 # freq de muestreo en Hz
l = 1 # longitud del intervalo 3-4 = 1s
M = Fm*l # cant de muestras
Pm = 1/Fm # período de muestreo

# vector de tiempos discretizados
t = np.arange(inicio, fin, Pm)

print(len(t)) # verificamos la cantidad de muestras

# nyquist de muestreo
Nm = Fm/2
print("Nyquist de muestreo: ", Nm, " [Hz]")

# funcion
x = (7* np.cos(40* np.pi * t)) + (5* np.cos(100* np.pi *t))

plt.figure(figsize=(16,8))
plt.plot(t,x)
plt.show()
```

$g(t)=7.sen(10\pi t)-8.cos(14\pi t)+ (1/2).sen(50\pi t)$
