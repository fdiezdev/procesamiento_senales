### **Librerías necesarias para realizar la tarea**

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.fftpack import fft,fftshift,ifft,fftfreq #funciones para trabajar con Transformada de Fourier
from scipy.io.wavfile import read, write  #para leer o grabar archivos de audio con formato wav
```

### **Para los 4 audios grabados se abordan todos los puntos pedidos:**

1) Llevar a memoria en Colab

2) Extraer Fr, la frecuencia de muestreo y guardar la señal en una variable.

```python
### AUDIO 1 ###
generador=read('1generador.wav')  #Cargo el archivo de audio "1generador.wav" en la variable generador

fgenerador=generador[0]     #Almaceno en la variable fgenerador el primer elemento de la variable generador =  frecuencia de muestreo
print('La frecuencia de muestreo del sonido "1generador" es '+ str(fgenerador) +' Hz') #Imprimo la frecuencia de muestreo

mat_generador=generador[1]  #Almaceno el segundo elemento de la variable generador, es decir, guardo la matriz del audio
sgenerador=mat_generador[:,0] #En la variable sgenerador guardo uno de los canales del audio, el de la columna 0

##El mismo código se repite para el resto de los audios de forma similar##

### AUDIO 2 ###
rueda=read('2rueda.wav')  

frueda = rueda[0]     
print('La frecuencia de muestreo del sonido "2rueda" es '+ str(frueda) +' Hz') 

mat_rueda=rueda[1]  
srueda=mat_rueda[:,0]


### AUDIO 3 ###
cuerda=read('3cuerda.wav')  

fcuerda = cuerda[0]     
print('La frecuencia de muestreo del sonido "3cuerda" es '+ str(fcuerda) +' Hz') 

mat_cuerda=cuerda[1]  
scuerda=mat_cuerda[:,0] 


### AUDIO 4 ###
monitor=read('4monitor.wav')  

fmonitor = monitor[0]    
print('La frecuencia de muestreo del sonido "4monitor" es '+ str(fmonitor) +' Hz') 

mat_monitor=monitor[1]  
smonitor=mat_monitor[:,0] 
```

3) Calcular la cantidad de muestras (M), la duración de la señal (en segundos) y definir el vector de tiempos discretizados.

4) Graficar la señal en función del tiempo discretizado.

```python
### AUDIO 1 ###
Mgenerador = len(sgenerador) #Calculo la cantidad de muestras del sonido "generador"
print('La cantidad de muestras del audio "1generador" es de '+ str(Mgenerador))
Tgenerador = Mgenerador/fgenerador  #Calculo el tiempo total del sonido "generador" medido en segundos. 
#Lo hago usando regla de 3 simple: fgenerador (cant de muestras en 1 segundo) ---- 1 segundo
                                  #Mgenerador (cant de muestras total) ------- Tgenerador (tiempo total) = Mgenerador*1s/fgenerador
print('El tiempo total del audio "1generador" es de '+ str(Tgenerador) + ' segundos' + '\n')

tdgenerador=np.arange(0,Mgenerador)/fgenerador #Genero el vector de tiempos discretizado del audio generador

plt.figure(figsize=(10,5))  #le doy un tamaño a la figura
plt.plot(tdgenerador,sgenerador, color='red') #grafico la señal del audio vs el tiempo y le otorgo un color de línea
plt.title('Audio 1: generador vs tiempo') #agrego un título al gráfico
plt.show() #muestro la figura y permito que se creen otras figuras separadas


##El mismo código se repite para el resto de los audios de forma similar##


### AUDIO 2 ###
Mrueda = len(srueda)
print('\n' + 'La cantidad de muestras del audio "2rueda" es de '+ str(Mrueda))
Trueda = Mrueda/frueda
print('El tiempo total del audio "2rueda" es de '+ str(Trueda) + ' segundos' + '\n')

tdrueda=np.arange(0,Mrueda)/frueda

plt.figure(figsize=(10,5))
plt.plot(tdrueda,srueda, color='blue')
plt.title('Audio 2: rueda vs tiempo')
plt.show()



### AUDIO 3 ###
Mcuerda = len(scuerda)
print('\n' + 'La cantidad de muestras del audio "3cuerda" es de '+ str(Mcuerda))
Tcuerda = Mcuerda/fcuerda
print('El tiempo total del audio "3cuerda" es de '+ str(Tcuerda) + ' segundos' + '\n')

tdcuerda=np.arange(0,Mcuerda)/fcuerda 

plt.figure(figsize=(10,5))
plt.plot(tdcuerda,scuerda, color='orange')
plt.title('Audio 3: cuerda vs tiempo')
plt.show()



### AUDIO 4 ###
Mmonitor = len(smonitor)
print('\n' + 'La cantidad de muestras del audio "4monitor" es de '+ str(Mmonitor))
Tmonitor = Mmonitor/fmonitor
print('El tiempo total del audio "4monitor" es de '+ str(Tmonitor) + ' segundos'+ '\n')

tdmonitor=np.arange(0,Mmonitor)/fmonitor 

plt.figure(figsize=(10,5))
plt.plot(tdmonitor,smonitor)
plt.title('Audio 4: monitor vs tiempo')
plt.show()
```

5) Calcular la Transformada de Fourier de la señal. (DFT punto N=M cantidad de muestras)

```python
### AUDIO 1 ###
TFgenerador=fft(sgenerador, Mgenerador) #calculo la DFT.N con N=Mgenerador (cantidad de muestras del audio generador)
fngenerador=fftfreq(Mgenerador,1/fgenerador) #creo el vector de frecuencias fundamentales en Hz (debido a que el período -1/fgenerador- está en segundos)

plt.figure(figsize=(8,5)) #defino el tamaño de la figura
plt.plot(fngenerador[0:Mgenerador//2],abs(TFgenerador[0:Mgenerador//2]),color='red') #grafico la mitad de la TF debido a que el resto de la información corresponde a la TF espejada. 
                                                                    #Las // indican que divido la variable Mgenerador (cantidad de muestras) a la mitad y me entrega un número entero.
                                                                    #El fngenerador es el vector de frecuencias fundamentales de la DFT.N creado más arriba
                                                                    #Queda graficada la TF respecto a las frecuencias en Hz
plt.title('Transformada de Fourier audio 1 generador vs frecuencia') #coloco un título a la figura
plt.show()

##El mismo código se repite para el resto de los audios de forma similar##

### AUDIO 2 ###
TFrueda=fft(srueda, Mrueda) 
fnrueda=fftfreq(Mrueda,1/frueda)

plt.figure(figsize=(8,5))
plt.plot(fnrueda[0:Mrueda//2],abs(TFrueda[0:Mrueda//2]),color='blue')
plt.title('Transformada de Fourier audio 2 rueda vs frecuencia')
plt.show()


### AUDIO 3 ###
TFcuerda=fft(scuerda, Mcuerda)
fncuerda=fftfreq(Mcuerda,1/fcuerda)

plt.figure(figsize=(7.83,5))
plt.plot(fncuerda[0:Mcuerda//2],abs(TFcuerda[0:Mcuerda//2]),color='orange')
plt.title('Transformada de Fourier audio 3 cuerda vs frecuencia')
plt.show()


### AUDIO 4 ###
TFmonitor=fft(smonitor, Mmonitor)
fnmonitor=fftfreq(Mmonitor,1/fmonitor)

plt.figure(figsize=(7.83,5))
plt.plot(fnmonitor[0:Mmonitor//2],abs(TFmonitor[0:Mmonitor//2]))
plt.title('Transformada de Fourier audio 4 monitor vs frecuencia')
plt.show()
```

6) Calcular la transformada de Fourier (punto N=Fr, es decir la cantidad de muestras  en 1s) en tres segundos diferentes de la señal. Analice los resultados de acuerdo a lo que escucha en los audios en esos momentos.

7) En cuanto a lo analizado en el ítem anterior, considera que la señal analizada es estacionaria o no estacionaria? Justifique.

```python
### AUDIO 1 ###

plt.figure(figsize=(20,8)) #defino el tamaño de cada figura

plt.subplot(231)  #utilizo el primer lugar del subplot de 2 filas y 3 columnas (primera fila, primera columna)
plt.plot(tdgenerador[0:fgenerador],sgenerador[0:fgenerador], color='red', label='Señal 1er segundo') #grafico el primer segundo de la señal generador vs su 
                                                                                              #vector de tiempos discretizado del primer segundo, se define trazo de color y leyenda
plt.legend(loc='upper right') #se define lugar de la leyenda en la figura
plt.title('1er segundo audio generador') #se asigna título a la figura

plt.subplot(232)  #utilizo el segundo lugar del subplot de 2 filas y 3 columnas (primera fila, segunda columna)
plt.plot(tdgenerador[fgenerador*2:fgenerador*3],sgenerador[fgenerador*2:fgenerador*3], color='brown', label='Señal 2do segundo') #comando similares para el 2do segundo
plt.legend(loc='upper right')
plt.title('2do segundo audio generador')

plt.subplot(233) #utilizo el tercer lugar del subplot de 2 filas y 3 columnas (primera fila, tercer columna)
plt.plot(tdgenerador[fgenerador*6:fgenerador*7],sgenerador[fgenerador*6:fgenerador*7], color='darkred', label='Señal 6to segundo') #comando similares para el 2do segundo
plt.legend(loc='upper right')
plt.title('6to segundo audio generador')

TF1secgenerador=fft(sgenerador[0:fgenerador], fgenerador) #calculo la DFT.N con N=frecuencia de muestreo para el 1er segundo
TF2secgenerador=fft(sgenerador[fgenerador*2:fgenerador*3], fgenerador)  #calculo la DFT.N con N=frecuencia de muestreo para el 2do segundo
TF6secgenerador=fft(sgenerador[fgenerador*6:fgenerador*7], fgenerador)  #calculo la DFT.N con N=frecuencia de muestreo para el 6to segundo
fnsecgenerador=fftfreq(fgenerador,1/fgenerador) #genero el vector de frecuencias fundamentales para la DFT.N con N=frecuencia de muestreo (fgenerador)

plt.subplot(234) #utilizo el cuarto lugar del subplot de 2 filas y 3 columnas (segunda fila, primera columna)
plt.plot(fnsecgenerador[0:fgenerador//2],abs(TF1secgenerador[0:fgenerador//2]),color='pink', label='TF 1er segundo') #grafico la mitad de la TF debido a que el resto de la 
              #información corresponde a la TF espejada. Las // indican que divido la variable fgenerador (cantidad de muestras de la DFT) a la mitad y me entrega un número entero
              #fnsecgenerador es el vector de frecuencias fundamentales creado más arriba
plt.legend(loc='upper right') #se define el lugar en la figura para la leyenda

plt.subplot(235) #utilizo el quinto lugar del subplot de 2 filas y 3 columnas (segunda fila, segunda columna), el resto del código es similar al usado para el 1er segundo
plt.plot(fnsecgenerador[0:fgenerador//2],abs(TF2secgenerador[0:fgenerador//2]),color='sandybrown', label='TF 2do segundo')
plt.legend(loc='upper right')

plt.subplot(236)#utilizo el sexto lugar del subplot de 2 filas y 3 columnas (segunda fila, tercer columna), el resto del código es similar al usado para el 1er segundo
plt.plot(fnsecgenerador[0:fgenerador//2],abs(TF6secgenerador[0:fgenerador//2]),color='gray', label='TF 6to segundo')
plt.legend(loc='upper right')

plt.show()
```

El audio 1generador.wav es una señal predominantemente estacionaria. Decimos esto porque salvo en el segundo 6 en el que claramente hay un cambio que se ve reflejado en los coeficientes de la transformada de Fourier de la señal, en términos generales, la señal es estacionaria. 

```python
##Para el resto de los audios se repite el código de forma similar##

### AUDIO 2 ###

plt.figure(figsize=(20,8))
plt.subplot(231)
plt.plot(tdrueda[0:frueda],srueda[0:frueda], color='blue', label='Señal 1er segundo')
plt.legend(loc='upper right')
plt.title('1er segundo audio rueda')
plt.subplot(232)
plt.plot(tdrueda[frueda*2:frueda*3],srueda[frueda*2:frueda*3], color='purple', label='Señal 2do segundo')
plt.legend(loc='upper right')
plt.title('2do segundo audio rueda')
plt.subplot(233)
plt.plot(tdrueda[frueda*6:frueda*7],srueda[frueda*6:frueda*7], color='deeppink', label='Señal 6to segundo')
plt.legend(loc='upper right')
plt.title('6to segundo audio rueda')

TF1secrueda=fft(srueda[0:frueda], frueda)
TF2secrueda=fft(srueda[frueda*2:frueda*3], frueda)
TF6secrueda=fft(srueda[frueda*6:frueda*7], frueda)
fnsecrueda=fftfreq(frueda,1/frueda)

plt.subplot(234)
plt.plot(fnsecrueda[0:frueda//2],abs(TF1secrueda[0:frueda//2]),color='cyan', label='TF 1er segundo')
plt.legend(loc='upper right')
plt.subplot(235)
plt.plot(fnsecrueda[0:frueda//2],abs(TF2secrueda[0:frueda//2]),color='darkviolet', label='TF 2do segundo')
plt.legend(loc='upper right')
plt.subplot(236)
plt.plot(fnsecrueda[0:frueda//2],abs(TF6secrueda[0:frueda//2]),color='k', label='TF 6to segundo')
plt.legend(loc='upper right')

plt.show()
```

El audio 2rueda.wav es una señal predominantemente estacionaria debido a que su transformada de Fourier en los 3 segundos elegidos es muy similar. Concluimos esto porque aunque al observar los gráficos de la transformada de Fourier en detalle, vemos que en el 6to segundo los coeficientes en frecuencias bajas son levemente menores y encontramos algunos picos más en frecuencias levemente mayores, la transformada de fourier es prácticamente la misma, por lo tanto, la señal en términos generales es estacionaria.

```python
### AUDIO 3 ###

plt.figure(figsize=(20,8))
plt.subplot(231)
plt.plot(tdcuerda[0:fcuerda],scuerda[0:fcuerda], color='orange', label='Señal 1er segundo')
plt.legend(loc='upper right')
plt.title('1er segundo audio cuerda')
plt.subplot(232)
plt.plot(tdcuerda[fcuerda*6:fcuerda*7],scuerda[fcuerda*6:fcuerda*7], color='y', label='Señal 6to segundo')
plt.legend(loc='upper right')
plt.title('6to segundo audio cuerda')
plt.subplot(233)
plt.plot(tdcuerda[fcuerda*16:fcuerda*17],scuerda[fcuerda*16:fcuerda*17], color='darkolivegreen', label='Señal 16vo segundo')
plt.legend(loc='upper right')
plt.title('16vo segundo audio cuerda')

TF1seccuerda=fft(scuerda[0:fcuerda], fcuerda)
TF6seccuerda=fft(scuerda[fcuerda*6:fcuerda*7], fcuerda)
TF16seccuerda=fft(scuerda[fcuerda*16:fcuerda*17], fcuerda)
fnseccuerda=fftfreq(fcuerda,1/fcuerda)

plt.subplot(234)
plt.plot(fnseccuerda[0:fcuerda//2],abs(TF1seccuerda[0:fcuerda//2]),color='tan', label='TF 1er segundo')
plt.legend(loc='upper right')
plt.subplot(235)
plt.plot(fnseccuerda[0:fcuerda//2],abs(TF6seccuerda[0:fcuerda//2]),color='gold', label='TF 6to segundo')
plt.legend(loc='upper right')
plt.subplot(236)
plt.plot(fnseccuerda[0:fcuerda//2],abs(TF16seccuerda[0:fcuerda//2]),color='greenyellow', label='TF 16vo segundo')
plt.legend(loc='upper right')

plt.show()
```

El audio 3cuerda.wav claramente es una señal NO estacionaria debido a que al escucharlo encontramos diferentes tonos de frecuencia a lo largo del tiempo y esto se ve reflejado en las transformadas de Fourier que son diferentes para los 3 segundos elegidos.

```python
### AUDIO 4 ###

plt.figure(figsize=(20,8))
plt.subplot(231)
plt.plot(tdmonitor[0:fmonitor],smonitor[0:fmonitor], label='Señal 1er segundo')
plt.legend(loc='upper right')
plt.title('1er segundo audio monitor')
plt.subplot(232)
plt.plot(tdmonitor[fmonitor*9:fmonitor*10],smonitor[fmonitor*9:fmonitor*10], color='cadetblue', label='Señal 9no segundo')
plt.legend(loc='upper right')
plt.title('9no segundo audio monitor')
plt.subplot(233)
plt.plot(tdmonitor[fmonitor*55:fmonitor*56],smonitor[fmonitor*55:fmonitor*56], color='green', label='Señal 55vo segundo')
plt.legend(loc='upper right')
plt.title('55vo segundo audio monitor')

TF1secmonitor=fft(smonitor[0:fmonitor], fmonitor)
TF9secmonitor=fft(smonitor[fmonitor*9:fmonitor*10], fmonitor)
TF55secmonitor=fft(smonitor[fmonitor*55:fmonitor*56], fmonitor)
fnsecmonitor=fftfreq(fmonitor,1/fmonitor)

plt.subplot(234)
plt.plot(fnsecmonitor[0:fmonitor//2],abs(TF1secmonitor[0:fmonitor//2]),color='darkblue', label='TF 1er segundo')
plt.legend(loc='upper right')
plt.subplot(235)
plt.plot(fnsecmonitor[0:fmonitor//2],abs(TF9secmonitor[0:fmonitor//2]),color='lightsteelblue', label='TF 9no segundo')
plt.legend(loc='upper right')
plt.subplot(236)
plt.plot(fnsecmonitor[0:fmonitor//2],abs(TF55secmonitor[0:fmonitor//2]),color='olive', label='TF 55vo segundo')
plt.legend(loc='upper right')

plt.show()
```

El audio 4monitor.wav es una señal NO estacionaria debido a que aunque parece tener los mismos componentes de Frecuencia los pesos de cada uno de los coeficientes varía a lo largo del tiempo, como se ve reflejado en los gráficos de los 3 segundos elegidos.
