## **Tipo de Imágenes**

Existen tres tipos principales de imágenes:

* **La imagen binaria:** es una imagen en blanco y negro. Cada píxel tiene asignado un valor lógico de 0 o 1.

* **La imagen de intensidad:** los elementos de una imagen de intensidad en general son de clase uint8 (enteros almacenados en 8 bits) o de clase uint16 (enteros almacenados en 16 bits), pueden almacenar, respectivamente, 2^8=256 valores en el rango [0-255], o 2^16=65536 valores en el rango [0-65535]. Cuando la imagen es de clase float32, los valores son números en punto flotante (que se almacenan en 32 bits)
    
* **La imagen en color:** es como la imagen de intensidad pero tiene tres canales, es decir, a cada píxel le corresponden tres valores de intensidad (formato RGB) en lugar de uno.

* **Imagen de categorías:** Puede visualizarse con colores.

# **Imágenes a color**

RGB: tres matrices monocromas (una para el rojo (R), una para el verde (G) y otra para el azul (B). Red, Green, Blue) se guardan en un arreglo de tamaño mxnx3 donde m es la cantidad de filas y n la de columnas, 3 son los canales (la tercera dimensión), es decir, una imagen a color de tamaño mxn.

a) Generar una imagen de color rojo.

```python
# Genero imagen de color rojo
R = np.zeros((256,256,3))
#la generamos llena de ceros, pero con las dimensiones necesarias aqui guardaremos la imagen , modificamos los valores:
R[:,:,0]=1;  #en la primera capa (que es la roja, le damos el máximo valor
#el resto de las capas quedaron 0 (sin encender)

#Graficamos esta imagen 
plt.imshow(R)
plt.title('Imagen RGB roja') 
plt.show()
print(R.dtype)
```

```python
# Genero imagen a color verde
G = np.zeros((256,256,3))
G[:,:,1]=1;  

#Graficamos esta imagen 
plt.imshow(G)
plt.title('Imagen RGB verde') 
plt.show()
```

```python
# Graficar una imagen (2n)x(2n)  dividida en cuatro colores: rojo, verde, azul y amarillo.

n=128; #elegimos un tamaño para n 
B=np.ones((n,n));
N=np.zeros((n,n));

roja = np.zeros((n,n,3))
verde = np.zeros((n,n,3))
azul = np.zeros((n,n,3))
amarillo = np.zeros((n,n,3))

roja[: , : , 0]= B;
roja[: , : , 1]= N;
roja[: , : , 2]= N;
 
verde[: , : , 0]= N;
verde[: , : , 1]= B;
verde[: , : , 2]= N;
 
azul[: , : , 0]= N;
azul[: , : , 1]= N;
azul[: , : , 2]= B;

amarillo[: , : , 0]=B;
amarillo[: , : , 1]=B;
amarillo[: , : , 2]=N;

Cuadro1= np.concatenate((roja, verde),axis=1)
Cuadro2= np.concatenate((azul, amarillo),axis=1)
Cuadro= np.concatenate((Cuadro1, Cuadro2),axis=0)

plt.imshow(Cuadro)
plt.title('Imagen generada concatenando matrices RGB (3D)') 
plt.show()
```
